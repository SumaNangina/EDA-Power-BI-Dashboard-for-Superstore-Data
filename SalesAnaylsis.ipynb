```python
import pandas as pd
df = pd.read_csv("cleaned_superstore.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.duplicated().sum()
```

```python
import matplotlib.pyplot as plt
category_sales = df.groupby('Category')['Total_Sales'].sum()
category_sales.plot(kind='bar', color='skyblue', title='Sales by Category')
plt.ylabel("Total Sales")
plt.xlabel("Category")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = df.resample('M', on='Order Date')['Total_Sales'].sum()
monthly_sales.plot(figsize=(10, 5), title="Monthly Sales Over Time")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
region_sales = df.groupby("Region")["Total_Sales"].sum()
region_sales.plot(kind="bar", color='lightgreen', title="Sales by Region")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
top_products = df.groupby("Product Name")["Total_Sales"].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='barh', color='orange', title="Top 10 Best-Selling Products")
plt.xlabel("Sales")
plt.gca().invert_yaxis()
plt.show()
```
